### Problem:

Objective: min x + x*y

Constraints:

-x + 2*y*x <= 8

2*x + y <= 14

2*x - y <= 10

0 <= x <= 10

0 <= y <= 10


```python
def f(x):

    pen = 0
    if not -x[0] +2*x[0]*x[1] <= 8:
        pen = np.inf
    if not 2*x[0] + x[1] <= 14:
        pen = np.inf
    if not 2*x[0] - x[1] <= 10:
        pen = np.inf
    return -(x[0]+x[1]*x[0]) + pen
```

```python
varbounds = np.array([[0,10],[0,10]])   

vartype = np.array([['int'],['real']])
```

```python
model = ga(function=f, dimension =2, variable_type_mixed=vartype, variable_boundaries=varbounds)
```

```python
model.run()
```
